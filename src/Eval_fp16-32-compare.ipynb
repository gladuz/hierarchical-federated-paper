# Plotting Figures for 16-bit vs 32-bit comparisons

```python
import pickle
import matplotlib.pyplot as plt
```

```python
# ===== MNIST MLP IID ===== 
datamodelset = "MNIST_MLP_IID FP16 and FP32 Comparison"
filename1 = "FL_mnist_mlp_468_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]_FP16"
filename2 = "HFL2_mnist_mlp_100_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]_FP16"
filename3 = "HFL4_mnist_mlp_100_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]_FP16"
filename4 = "HFL4_mnist_mlp_30_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]_FP16"

filename5 = "[1]FL_mnist_mlp_468_C[0.1]_iid[1]_E[1]_B[10]"
filename6 = "[3]HFL2_mnist_mlp_100_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]"
filename7 = "[5]HFL4_mnist_mlp_100_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]"
filename8 = "[7]HFL4_mnist_mlp_30_lr[0.01]_C[0.1]_iid[1]_E[1]_B[10]"

with open(r"../save/objects_fp16/" + filename1 + ".pkl", "rb") as input_file: data1 = pickle.load(input_file)
with open(r"../save/objects_fp16/" + filename2 + ".pkl", "rb") as input_file: data2 = pickle.load(input_file)
with open(r"../save/objects_fp16/" + filename3 + ".pkl", "rb") as input_file: data3 = pickle.load(input_file)
with open(r"../save/objects_fp16/" + filename4 + ".pkl", "rb") as input_file: data4 = pickle.load(input_file)    

with open(r"../save/objects/" + filename5 + ".pkl", "rb") as input_file: data5 = pickle.load(input_file)
with open(r"../save/objects/" + filename6 + ".pkl", "rb") as input_file: data6 = pickle.load(input_file)
with open(r"../save/objects/" + filename7 + ".pkl", "rb") as input_file: data7 = pickle.load(input_file)
with open(r"../save/objects/" + filename8 + ".pkl", "rb") as input_file: data8 = pickle.load(input_file)    

trloss1 = data1[0]    
trloss2 = data2[0]    
trloss3 = data3[0]    
trloss4 = data4[0]    

trloss5 = data5[0]    
trloss6 = data6[0]    
trloss7 = data7[0]    
trloss8 = data8[0] 

tracc1 = data1[1]
tracc2 = data2[1]
tracc3 = data3[1]
tracc4 = data4[1]

tracc5 = data5[1]
tracc6 = data6[1]
tracc7 = data7[1]
tracc8 = data8[1]
```

```python
# PLOTTING 

# Plot Average Accuracy vs Communication rounds
plt.figure()
plt.title(datamodelset)

plt.plot(range(len(tracc1)), tracc1, ':', label="FL-FP16", linewidth=0.9)
plt.plot(range(len(tracc2)), tracc2, ':', label="HFL2-FP16", linewidth=0.9)
plt.plot(range(len(tracc3)), tracc3, ':', label="HFL4-FP16", linewidth=0.9)
plt.plot(range(len(tracc4)), tracc4, ':', label="HFL8-FP16")

plt.plot(range(len(tracc5)), tracc5, label="FL", linewidth=0.9)
plt.plot(range(len(tracc6)), tracc6, label="HFL2", linewidth=0.9)
plt.plot(range(len(tracc7)), tracc7, label="HFL4", linewidth=0.9)
plt.plot(range(len(tracc8)), tracc8, label="HFL8")
plt.legend(loc="lower right")

plt.ylabel('Average Accuracy')
plt.xlabel('Communication Rounds')
plt.savefig('../save/' + datamodelset + '_acc_FP16_32.png')

# Plot Loss curve
plt.figure()
plt.title(datamodelset)

plt.plot(range(len(trloss1)), trloss1, ':', label="FP-FP16", linewidth=0.9)
plt.plot(range(len(trloss2)), trloss2, ':', label="HFL2-FP16", linewidth=0.9)
plt.plot(range(len(trloss3)), trloss3, ':', label="HFL4-FP16", linewidth=0.9)
plt.plot(range(len(trloss4)), trloss4, ':', label="HFL8-FP16")

plt.plot(range(len(trloss5)), trloss5, label="FL", linewidth=0.9)
plt.plot(range(len(trloss6)), trloss6, label="HFL2", linewidth=0.9)
plt.plot(range(len(trloss7)), trloss7, label="HFL4", linewidth=0.9)
plt.plot(range(len(trloss8)), trloss8, label="HFL8")
plt.legend(loc="upper right")

plt.ylabel('Training loss')
plt.xlabel('Communication Rounds')
plt.savefig('../save/' + datamodelset + '_loss_FP16_32.png')

plt.show()
```
